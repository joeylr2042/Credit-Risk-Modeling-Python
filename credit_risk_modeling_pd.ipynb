# Data Preparation: An Example

```python
df_x_prepr = loan_data_x_train
df_y_prepr = loan_data_y_train
```

```python
df_x_prepr['grade'] .unique()
```

```python
df1 = pd.concat([df_x_prepr['grade'], df_y_prepr], axis = 1)
df1.head()
```

```python
df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].sum()
```

```python
df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].mean()
```

```python
df1 = pd.concat([df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].sum(), df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].mean()], axis = 1)
df1
```

```python
df1 = df1.iloc[:, [0, 1, 3]]
df1
```

```python
df1.columns = [df1.columns.values[0], 'n_obs', 'prop_good']
df1
```

```python
df1['prop_n_obs'] = df1['n_obs'] / df1['n_obs'].sum()
df1
```

```python
df1['n_good'] = df1['prop_good'] * df1['n_obs']
df1['n_bad'] = (1 - df1['prop_good']) * df1['n_obs']
df1
```

```python
df1['prop_n_good'] = df1['n_good'] / df1['n_good'].sum()
df1['prop_n_bad'] = df1['n_bad'] / df1['n_bad'].sum()
df1
```

```python
df1['WoE'] = np.log(df1['prop_n_good'] / df1['prop_n_bad'])
df1
```

```python
df1 = df1.sort_values(['WoE'])
df1 = df1.reset_index(drop = True)
df1
```

```python
df1['diff_prop_good'] = df1['prop_good'].diff().abs()
df1['diff_WoE'] = df1['WoE'].diff().abs()
df1
```

```python
df1['IV'] = (df1['prop_n_good'] - df1['prop_n_bad']) * df1['WoE']
df1['IV'] = df1['IV'].sum()
df1
```
